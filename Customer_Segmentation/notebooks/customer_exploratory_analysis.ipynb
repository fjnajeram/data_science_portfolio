## Exploratory analysis of customer data

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# read in the data
data_path = '../data/sample_customers.csv'
df = pd.read_csv(data_path)
# drop unnamed column
df.drop(columns=['Unnamed: 0'], inplace=True)
# describe the data including categorical and numerical variables
df.describe(include='all')
```

```python
df.dtypes
```

```python
# histograms for numerical data. Exclude column 0
df.drop(columns=['CustomerID']).hist(figsize=(15,15))
```

```python
# bar plot distributions for categorical data
cat_vars = ['Gender', 'Location']
for var in cat_vars:
    sns.countplot(x=var, data=df)
    plt.xticks(rotation=90)
    plt.show()
```
